```python
# Naive Bayes

import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
```

```python
#importing the dataset
dataset = pd.read_table('iri', sep=',')
print(dataset)

print ("Shape of the data:",dataset.shape,"\n")
```

```python
#result of independent variable
X = dataset.iloc[:,0:8].values
print("\n Result of independent variable = X:\n",X,"\n")
```

```python
#result of dependent variable 
y = dataset.iloc[:, -1].values
print("\n dependent variable = y = ",y,"\n")
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()
X[:,0]= labelencoder_X.fit_transform(X[:,0])
print("\n",X,"\n")

print ("Shape of the data:",dataset.shape,"\n")

X = np.array(X)
y = np.array(y)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 20)

model = GaussianNB()
model.fit(X_train,y_train)
predicted1 = model.predict(X_test)
print("Result\n",predicted1)
```
